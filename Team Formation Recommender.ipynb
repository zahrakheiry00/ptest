```python
import pandas as pd
datafile_path = 'dblp_papers_v11.txt'


chunk_size = 100000
for chunk in pd.read_json(datafile_path, lines = True, chunksize = chunk_size):
    aggregation = list()
    for index, entry in chunk.iterrows():
        id_value = entry['id']
        element_value = entry['authors']
        if entry['fos'] != 'nan': 
            keyword_value = entry['fos']
        else: 
            print('nan')
            keyword_value = []
        aggregation.append({'id': id_value, 'authors': len(list(element_value)), 'keywords': keyword_value})
        
    aggregation_df = pd.DataFrame(aggregation)
    aggregation_df = aggregation_df.dropna()
    aggregation_df['skills'] = aggregation_df.apply(lambda row : len(list(row['keywords'])), axis = 1)
    aggregation_df = aggregation_df[['id', 'authors', 'skills']]
    aggregation_df.to_csv('my_csv_output.csv', mode = 'a', header = False, index = False)
```

```python
import pandas as pd
file_path = 'my_csv_output.csv'
data = pd.read_csv(file_path)
column_names = ['ID', 'num_authors', 'num_skills']
data = pd.read_csv(file_path, names= column_names)
print(data.head())
```
